```python
import pandas as pd;
import numpy as np;
import matplotlib.pyplot as plt;
from sklearn.model_selection import train_test_split;
from sklearn import tree
```

```python
df = pd.read_csv("titanic.csv")
df=df.drop(["PassengerId","Name","SibSp","Parch","Cabin","Embarked","Ticket"],axis=1)
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df["Sex"] = le.fit_transform(df["Sex"])
df.head()
```

```python
x = df.drop("Survived",axis=1)
y = df["Survived"]
x_train,x_test,y_train,y_test = train_test_split(x,y)
```

```python
model = tree.DecisionTreeClassifier(max_depth = 4,random_state = 42)
```

```python
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
model.score(x_train,y_train)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(30, 18))  # Adjust figure size
plot_tree(
    model, 
    feature_names=x.columns,   # Use DataFrame column names as feature names
    filled=True,              # Fill with colors for clarity
    rounded=True,             # Use rounded edges
    fontsize=10               # Font size for text
)
plt.axis('off') 
```
