```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import DataLoader
from torchvision.datasets import MNIST
import torchvision.transforms as transforms
```

```python
class NN(nn.Module):
  def __init__(self,input_size,num_classes):
    super(NN,self).__init__()
    self.fc1 = nn.Linear(input_size,50)
    self.fc2 = nn.Linear(50,100)
    self.fc3 = nn.Linear(100,num_classes)

  def forward(self,X):
    fc1_output = F.relu(self.fc1(X))
    fc2_output = F.relu(self.fc2(fc1_output))
    fc3_output = self.fc3(fc2_output)
    return fc3_output
```

```python
input_size = 784
num_classes = 10
epochs = 1
lr = 0.001
batch_size = 128
```

```python
trainData = MNIST(root="dataset/",train=True,transform = transforms.ToTensor(),download=True)
train_batches = DataLoader(dataset = trainData,batch_size=batch_size,shuffle = True)
testData = MNIST(root="dataset/",train=False,transform = transforms.ToTensor(),download=True)
test_batches = DataLoader(dataset = testData,batch_size=batch_size,shuffle = True)
```

```python
import torch.cuda
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
model = NN(input_size=input_size,num_classes=num_classes)
```

```python
creterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(),lr=lr)
```

```python
for i in range(epochs):
  for batch_idx ,(data,targets) in enumerate(train_batches):
    data = data.to(device)
    targets = targets.to(device)

    data = data.reshape(data.shape[0],-1)

    scores= model(data)
    loss = creterion(scores,targets)

    optimizer.zero_grad()
    loss.backward()

    optimizer.step()
```

```python
model.eval()
```

```python
def check_accuracy(loader_data,model):
  num_sam = 0
  num_crt = 0

  model.eval()
  with torch.no_grad():
    for x,y in (loader_data):
      x = x.to(device)
      y = y.to(device)

      x = x.reshape(x.shape[0],-1)

      scores = model(x)
      _, predict = scores.max(1)
      num_crt += (predict == y).sum()
      num_sam += len(y)

    model.train()
    return num_crt/num_sam
```

```python
print(check_accuracy(test_batches,model))
```

```python
def predict(loader_data,model):
  output=[]
  model.eval()
  with torch.no_grad():
    for x,y in (loader_data):
      x = x.to(device)
      y = y.to(device)

      x = x.reshape(x.shape[0],-1)

      scores = model(x)
      _, predict = scores.max(1)
      output.append(list(predict))
    model.train()
    return output
```

```python
print(predict(test_batches,model))
```
